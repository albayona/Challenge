# 3. Normalization and K-means Clustering Summary

This notebook summarizes the data normalization and K-means clustering process for the stock recommendation dataset.

## Overview
The dataset contains independent features (engineered features + reduced technical indicators) ready for clustering analysis. We normalize the data and perform K-means clustering to identify patterns in stock recommendations.

## Data Sources
- **Independent Features Dataset**:  - Features ready for clustering
- **Process**: Load → Data Quality Check → Normalization → Optimal K determination → K-means clustering → Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings("ignore")

plt.style.use("default")
sns.set_palette("husl")
print("Libraries imported successfully!")
```

## 1. Load Independent Features Dataset

```python


# Load the CSV
df = pd.read_csv("stock_data_independent_features.csv")

# Replace Inf and -Inf with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop all rows with any NaN values
df_clean = df.dropna()

# Print shape difference
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_clean.shape}")
print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")
```

## 2. Data Normalization

### Normalization Methods
- **MinMaxScaler**: Scales features to [0,1] range - preserves original distribution shape
- **StandardScaler**: Z-score normalization (mean=0, std=1) - assumes normal distribution

Both methods ensure all features contribute equally to clustering by removing scale differences.

```python
# Apply both normalization methods
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# Normalize the data
df_minmax = pd.DataFrame(minmax_scaler.fit_transform(df_clean), columns=df_clean.columns)
df_standard = pd.DataFrame(standard_scaler.fit_transform(df_clean), columns=df_clean.columns)

print(f"NaN values df_minmax: {df_minmax.isnull().sum().sum()}")

print(f"NaN values df_standard: {df_standard.isnull().sum().sum()}")

print("=== NORMALIZATION RESULTS ===")
print(f"MinMaxScaler - Range: [{df_minmax.min().min():.3f}, {df_minmax.max().max():.3f}]")
print(f"StandardScaler - Mean: {df_standard.mean().mean():.3f}, Std: {df_standard.std().mean():.3f}")

# Show statistics for both methods
print(df_minmax.describe().round(3))
print(df_standard.describe().round(3))
```

## 3. Optimal Cluster Determination

### Methods Used
- **Elbow Method**: Plot inertia vs K to find the "elbow" point
- **Silhouette Analysis**: Measure cluster quality (higher is better, range: -1 to 1)

We test both normalization methods to find the best approach for our data.

```python
# Test different K values for both normalization methods
k_range = range(2, 11)
inertias_minmax = []
inertias_standard = []
silhouettes_minmax = []
silhouettes_standard = []

print("Testing K values from 2 to 10...")

for k in k_range:
    # MinMaxScaler
    kmeans_minmax = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_minmax.fit(df_minmax)
    inertias_minmax.append(kmeans_minmax.inertia_)
    silhouettes_minmax.append(silhouette_score(df_minmax, kmeans_minmax.labels_))
    
    # StandardScaler
    kmeans_standard = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_standard.fit(df_standard)
    inertias_standard.append(kmeans_standard.inertia_)
    silhouettes_standard.append(silhouette_score(df_standard, kmeans_standard.labels_))

print("Testing completed!")

# Find optimal K for each method
optimal_k_minmax = k_range[np.argmax(silhouettes_minmax)]
optimal_k_standard = k_range[np.argmax(silhouettes_standard)]

print(f"Optimal K - MinMaxScaler: {optimal_k_minmax} (silhouette: {max(silhouettes_minmax):.3f})")
print(f"Optimal K - StandardScaler: {optimal_k_standard} (silhouette: {max(silhouettes_standard):.3f})")
```

```python
# Plot results for both normalization methods
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# Elbow plots
ax1.plot(k_range, inertias_minmax, "bo-", linewidth=2, markersize=8)
ax1.set_title("Elbow Method - MinMaxScaler")
ax1.set_xlabel("Number of Clusters (K)")
ax1.set_ylabel("Inertia")
ax1.grid(True, alpha=0.3)

ax2.plot(k_range, inertias_standard, "ro-", linewidth=2, markersize=8)
ax2.set_title("Elbow Method - StandardScaler")
ax2.set_xlabel("Number of Clusters (K)")
ax2.set_ylabel("Inertia")
ax2.grid(True, alpha=0.3)

# Silhouette plots
ax3.plot(k_range, silhouettes_minmax, "bo-", linewidth=2, markersize=8)
ax3.set_title("Silhouette Score - MinMaxScaler")
ax3.set_xlabel("Number of Clusters (K)")
ax3.set_ylabel("Silhouette Score")
ax3.grid(True, alpha=0.3)

ax4.plot(k_range, silhouettes_standard, "ro-", linewidth=2, markersize=8)
ax4.set_title("Silhouette Score - StandardScaler")
ax4.set_xlabel("Number of Clusters (K)")
ax4.set_ylabel("Silhouette Score")
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. K-means Clustering Implementation

### Final Clustering
Based on silhouette scores, we select the best normalization method and optimal K for final clustering. The results are visualized using PCA for 2D representation.

```python
# === FINAL CLUSTERING FOR BOTH METHODS ===

# MinMaxScaler-based clustering
kmeans_minmax = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels_minmax = kmeans_minmax.fit_predict(df_minmax)

df_clusters_minmax = df_clean.copy()
df_clusters_minmax["cluster_minmax"] = cluster_labels_minmax

print("=== MinMaxScaler Clustering ===")
print(f"Optimal K: {optimal_k_minmax}")
print(f"Best silhouette score: {max(silhouettes_minmax):.3f}")
print("Cluster distribution (MinMax):")
print(df_clusters_minmax["cluster_minmax"].value_counts().sort_index())
print("\n")

# StandardScaler-based clustering
kmeans_standard = KMeans(n_clusters=optimal_k_standard, random_state=42, n_init=10)
cluster_labels_standard = kmeans_standard.fit_predict(df_standard)

df_clusters_standard = df_clean.copy()
df_clusters_standard["cluster_standard"] = cluster_labels_standard

print("=== StandardScaler Clustering ===")
print(f"Optimal K: {optimal_k_standard}")
print(f"Best silhouette score: {max(silhouettes_standard):.3f}")
print("Cluster distribution (Standard):")
print(df_clusters_standard["cluster_standard"].value_counts().sort_index())
print("\n")

# Merge both clustering results for comparison
df_with_clusters = df_clean.copy()
df_with_clusters["cluster_minmax"] = cluster_labels_minmax
df_with_clusters["cluster_standard"] = cluster_labels_standard

print("Clustering completed for both scaling methods.")
print(f"Final dataset shape: {df_with_clusters.shape}")
```
