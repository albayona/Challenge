# Correlation Analysis and Feature Reduction
## Remove redundant features and create independent feature set


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
print("Libraries imported successfully!")
```

```python
# Load data and drop established redundant columns
df = pd.read_csv('stock_data_final_clean.csv')

# Drop columns we already established as redundant
cols_to_drop = [
    'ticker', 'company',  # Identifiers
    'time', 'date', 'year', 'month', 'day', 'weekday', 'weekday_num', 'quarter',  # Time columns
     'rating_to', 'rating_from',  # Drop one from each highly correlated pair
    'action'  # Categorical action column
]

df_analysis = df.drop(columns=cols_to_drop)
print(f"Original shape: {df.shape}")
print(f"After dropping redundant columns: {df_analysis.shape}")
print(f"Remaining columns: {list(df_analysis.columns)}")
```

```python
# Calculate correlation matrix and find highly correlated pairs
correlation_matrix = df_analysis.corr()

# Find highly correlated pairs (absolute correlation > 0.8)
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > 0.8:
            high_corr_pairs.append({
                'feature1': correlation_matrix.columns[i],
                'feature2': correlation_matrix.columns[j],
                'correlation': corr_val
            })

print(f"Found {len(high_corr_pairs)} highly correlated pairs (|correlation| > 0.8):")
for pair in sorted(high_corr_pairs, key=lambda x: abs(x['correlation']), reverse=True):
    print(f"  {pair['feature1']} ↔ {pair['feature2']}: {pair['correlation']:.3f}")
```

```python
# First heatmap - showing all correlations
plt.figure(figsize=(15, 12))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(
    correlation_matrix, 
    mask=mask,
    annot=True, 
    cmap='RdBu_r', 
    center=0,
    square=True,
    fmt='.2f',
    cbar_kws={"shrink": .8}
)
plt.title('First Heatmap - All Features Correlations', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Group highly correlated features and keep only one from each group
correlation_groups = []
used_features = set()

for pair in sorted(high_corr_pairs, key=lambda x: abs(x['correlation']), reverse=True):
    f1, f2 = pair['feature1'], pair['feature2']
    
    # Skip if either feature is already grouped
    if f1 in used_features or f2 in used_features:
        continue
    
    # Create a new group
    group = {
        'features': [f1, f2],
        'max_correlation': abs(pair['correlation']),
        'correlations': [pair['correlation']]
    }
    
    # Add more features to this group if they're highly correlated with existing members
    for other_pair in high_corr_pairs:  
        
        if other_pair['feature1'] in group['features'] and other_pair['feature2'] not in used_features:
            group['features'].append(other_pair['feature2'])
            group['correlations'].append(other_pair['correlation'])
            used_features.add(other_pair['feature2'])
        elif other_pair['feature2'] in group['features'] and other_pair['feature1'] not in used_features:
            group['features'].append(other_pair['feature1'])
            group['correlations'].append(other_pair['correlation'])
            used_features.add(other_pair['feature1'])
    
    used_features.update(group['features'])
    correlation_groups.append(group)

print(f"Created {len(correlation_groups)} correlation groups:")
for i, group in enumerate(correlation_groups, 1):
    print(f"\nGroup {i}: {group['features']}")
    print(f"  - Keep: {group['features'][0]} (representative feature)")
    print(f"  - Drop: {group['features'][1:]}")
    print(f"  - Max correlation: {group['max_correlation']:.3f}")
```

```python
# Create reduced dataset with independent features
reduced_features = list(df_analysis.columns)

# Remove redundant features from each correlation group
for i, group in enumerate(correlation_groups):
    if i == 2:
        continue
    # Keep the first feature, drop the rest
    features_to_remove = group['features'][1:]
    for feature in features_to_remove:
        if feature in reduced_features:
            reduced_features.remove(feature)

print(f"Reduced from {len(df_analysis.columns)} to {len(reduced_features)} features")
print(f"\nRemoved features:")
removed_features = [col for col in df_analysis.columns if col not in reduced_features]
for feature in removed_features:
    print(f"  - {feature}")

# Create the reduced dataset
df_reduced = df_analysis[reduced_features]
print(f"\nReduced dataset shape: {df_reduced.shape}")
print(f"\nRemaining independent features: {list(df_reduced.columns)}")
```

```python
# Second heatmap - showing independent features correlations
correlation_matrix_reduced = df_reduced.corr()

plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix_reduced, dtype=bool))
sns.heatmap(
    correlation_matrix_reduced, 
    mask=mask,
    annot=True, 
    cmap='RdBu_r', 
    center=0,
    square=True,
    fmt='.2f',
    cbar_kws={"shrink": .8}
)
plt.title('Second Heatmap - Independent Features Correlations', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Verify independence - check for any remaining high correlations
remaining_high_corr = []
for i in range(len(correlation_matrix_reduced.columns)):
    for j in range(i+1, len(correlation_matrix_reduced.columns)):
        corr_val = correlation_matrix_reduced.iloc[i, j]
        if abs(corr_val) > 0.8:
            remaining_high_corr.append({
                'feature1': correlation_matrix_reduced.columns[i],
                'feature2': correlation_matrix_reduced.columns[j],
                'correlation': corr_val
            })

print(f"\nRemaining high correlations (|correlation| > 0.8): {len(remaining_high_corr)}")
if remaining_high_corr:
    for pair in remaining_high_corr:
        print(f"  {pair['feature1']} ↔ {pair['feature2']}: {pair['correlation']:.3f}")
else:
    print("✅ All features are now independent!")
```

```python
# Save the reduced dataset
df_reduced.to_csv('stock_data_independent_features.csv', index=False)
print(f"✅ Independent features dataset saved as 'stock_data_independent_features.csv'")

# Summary
print(f"\n=== SUMMARY ===")
print(f"Original features: {len(df.columns)}")
print(f"After dropping established redundant columns: {len(df_analysis.columns)}")
print(f"After removing highly correlated features: {len(df_reduced.columns)}")
print(f"Total reduction: {len(df.columns) - len(df_reduced.columns)} features")
print(f"Reduction percentage: {((len(df.columns) - len(df_reduced.columns)) / len(df.columns)) * 100:.1f}%")
```
