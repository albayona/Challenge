## **1. Import Libraries**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
# # Use a compatible matplotlib style
plt.style.use("seaborn-v0_8")  # Updated style name
import seaborn as sns
from ydata_profiling import ProfileReport

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors
from tabulate import tabulate
from kneed import KneeLocator

import warnings
warnings.filterwarnings('ignore')
```

## **1. Data Understanding**

First, we load the data and verify that it has loaded correctly

```python
stock_data = pd.read_csv("extracted_stock_data.csv")
stock_data["time"] = pd.to_datetime(stock_data["time"], errors="coerce")
```

## **2. Data Preparation**

After profiling, we analyze the columns and prepare the data for analysis.

```python
# Check for missing values
missing_data = stock_data.isnull().sum()
print("Missing values per column:")
print(missing_data[missing_data > 0])

if missing_data.sum() > 0:
    print(f"\nTotal missing values: {missing_data.sum()}")
    print("Consider removing rows with missing values or imputing them")
else:
    print("\n✓ No missing values found")
```

```python
# Check for duplicates
duplicates = stock_data.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

if duplicates > 0:
    print("Consider removing duplicate rows")
else:
    print("✓ No duplicate rows found")
```

## **3. Categorical Variables Analysis**

```python
categorical_cols = stock_data.select_dtypes(include=['object']).columns
print(f"Categorical columns: {list(categorical_cols)}")

for col in categorical_cols:
    print(f"\n{col.upper()}:")
    print("-" * (len(col) + 1))
    value_counts = stock_data[col].value_counts()
    print(f"Unique values: {stock_data[col].nunique()}")
    print(f"Most frequent: {value_counts.index[0]} ({value_counts.iloc[0]} times)")
    print(f"Top 5 values:")
    print(value_counts.head())
```

### Categorical Variables Visualization

```python
# Plot categorical distributions
for col in categorical_cols:
    if stock_data[col].nunique() <= 20:  # Only plot if not too many categories
        plt.figure(figsize=(12, 6))
        stock_data[col].value_counts().plot(kind='bar')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

## **4. Numerical Variables Analysis**

```python
numerical_cols = stock_data.select_dtypes(include=[np.number]).columns
print(f"Numerical columns: {list(numerical_cols)}")

if len(numerical_cols) > 0:
    print("\nDescriptive Statistics:")
    print("-" * 25)
    print(stock_data[numerical_cols].describe())
else:
    print("No numerical columns found.")
```

### Numerical Variables Visualization

```python
# Visualize numerical distributions
if len(numerical_cols) > 0:
    n_cols = min(3, len(numerical_cols))
    n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    if n_rows == 1:
        axes = [axes] if n_cols == 1 else axes
    else:
        axes = axes.flatten()
    
    for i, col in enumerate(numerical_cols):
        if i < len(axes):
            stock_data[col].hist(bins=30, ax=axes[i])
            axes[i].set_title(f'Distribution of {col}')
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('Frequency')
    
    # Hide empty subplots
    for i in range(len(numerical_cols), len(axes)):
        axes[i].set_visible(False)
    
    plt.tight_layout()
    plt.show()
```

## **5. Time Series Analysis**

```python
# Check if there's a time column
time_cols = [col for col in stock_data.columns if 'time' in col.lower() or 'date' in col.lower()]
print(f"Time-related columns found: {time_cols}")

if time_cols:
    time_col = time_cols[0]
    print(f"\nAnalyzing time column: {time_col}")
    
    # Convert to datetime if possible
    try:
        stock_data[time_col] = pd.to_datetime(stock_data[time_col])
        print(f"✓ Successfully converted {time_col} to datetime")
        
        # Time range analysis
        print(f"\nTime Range:")
        print(f"Earliest: {stock_data[time_col].min()}")
        print(f"Latest: {stock_data[time_col].max()}")
        print(f"Duration: {stock_data[time_col].max() - stock_data[time_col].min()}")
        
    except Exception as e:
        print(f"Could not analyze time patterns: {e}")
else:
    print("No time-related columns found.")
```

### Time Patterns Visualization

```python
if time_cols:
    time_col = time_cols[0]
    try:
        # Time-based aggregations
        stock_data['year'] = stock_data[time_col].dt.year
        stock_data['month'] = stock_data[time_col].dt.month
        stock_data['day_of_week'] = stock_data[time_col].dt.day_name()
        
        print(f"\nRecords by Year:")
        print(stock_data['year'].value_counts().sort_index())
        
        print(f"\nRecords by Month:")
        print(stock_data['month'].value_counts().sort_index())
        
        print(f"\nRecords by Day of Week:")
        print(stock_data['day_of_week'].value_counts())
        
        # Visualize time patterns
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
        # Yearly distribution
        stock_data['year'].value_counts().sort_index().plot(kind='bar', ax=axes[0,0])
        axes[0,0].set_title('Records by Year')
        axes[0,0].set_xlabel('Year')
        axes[0,0].set_ylabel('Count')
        
        # Monthly distribution
        stock_data['month'].value_counts().sort_index().plot(kind='bar', ax=axes[0,1])
        axes[0,1].set_title('Records by Month')
        axes[0,1].set_xlabel('Month')
        axes[0,1].set_ylabel('Count')
        
        # Day of week distribution
        day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        stock_data['day_of_week'].value_counts().reindex(day_order).plot(kind='bar', ax=axes[1,0])
        axes[1,0].set_title('Records by Day of Week')
        axes[1,0].set_xlabel('Day of Week')
        axes[1,0].set_ylabel('Count')
        axes[1,0].tick_params(axis='x', rotation=45)
        
        # Time series plot
        daily_counts = stock_data.set_index(time_col).resample('D').size()
        daily_counts.plot(ax=axes[1,1])
        axes[1,1].set_title('Daily Record Count Over Time')
        axes[1,1].set_xlabel('Date')
        axes[1,1].set_ylabel('Count')
        
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"Could not create time visualizations: {e}")
```

## **6. Correlation Analysis**

```python
numerical_cols = stock_data.select_dtypes(include=[np.number]).columns

if len(numerical_cols) > 1:
    correlation_matrix = stock_data[numerical_cols].corr()
    
    print("Correlation Matrix:")
    print(correlation_matrix.round(3))
    
    # Visualize correlation matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
               square=True, fmt='.2f')
    plt.title('Correlation Matrix of Numerical Variables')
    plt.tight_layout()
    plt.show()
else:
    print("Not enough numerical variables for correlation analysis.")
```

## **7. Data Quality Summary**

```python
print("DATASET OVERVIEW:")
print(f"  • Total Records: {len(stock_data):,}")
print(f"  • Total Columns: {len(stock_data.columns)}")
print(f"  • Memory Usage: {stock_data.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

print(f"\nCOLUMN SUMMARY:")
for col in stock_data.columns:
    dtype = stock_data[col].dtype
    null_count = stock_data[col].isnull().sum()
    null_pct = (null_count / len(stock_data)) * 100
    unique_count = stock_data[col].nunique()
    
    print(f"  • {col}: {dtype}, {null_count} nulls ({null_pct:.1f}%), {unique_count} unique values")

print(f"\nDATA QUALITY SCORE: {((len(stock_data) - stock_data.isnull().sum().sum() - stock_data.duplicated().sum()) / len(stock_data)) * 100:.1f}%")
```
